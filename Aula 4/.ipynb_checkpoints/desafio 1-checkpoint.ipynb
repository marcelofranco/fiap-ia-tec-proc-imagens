```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread("imagens/playing-cards.jpg", cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,10))
plt.imshow(image)
plt.title("Cartas")

image_cards = cv2.imread("imagens/playing-cards-7-copas.jpg", cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,10))
plt.imshow(image_cards)
plt.title("Cartas")
```

```python
imagem_correspondente = image.copy()

correspondencia_template = cv2.matchTemplate(image, image_cards, cv2.TM_CCOEFF_NORMED)
_, w, h = image_cards.shape[::-1]

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(correspondencia_template)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(imagem_correspondente, top_left, bottom_right, (0,255,0), 2)

plt.figure(figsize=(20,20))
plt.imshow(imagem_correspondente)
plt.title("7 de espada")
```

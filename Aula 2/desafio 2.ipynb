```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

%matplotlib inline

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
imagem = cv2.imread("imagens/gize.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(imagem)
```

```python
imagem_piramide = imagem.copy()

color = (0, 0, 255)

cv2.circle(imagem_piramide, (290,160), 5, color, 3)
cv2.circle(imagem_piramide, (530,425), 5, color, 3)
cv2.circle(imagem_piramide, (55,405), 5, color, 3)

plt.figure(figsize=(20,10))
plt.imshow(imagem_piramide)
plt.title("Piramide com marcadores")
```

```python
height, width = imagem.shape[:2]


mascara = np.zeros((height,width,3), np.uint8)

mascara[:,:] = (0,255,0)

#pontos = np.array([[0,0],[width,0],[0,height],[width,height]], np.int32)

#cv2.fillPoly(mascara, [pontos], 255)

#mascara.setTo(cv2.Scalar(0,255,0))

pontos = np.array([[290,160], [530,425], [55,405]], np.int32)
cv2.fillPoly(mascara, [pontos], (255,255,255))

plt.figure(figsize=(20,10))
plt.imshow(mascara)
plt.title("Mscara Pirmide")
```

```python
imagem_extraida = cv2.bitwise_and(imagem,mascara)

plt.figure(figsize=(20,10))
plt.imshow(imagem_extraida)
plt.title("Imagem Extra'dia")
```

```python
imagem_extraida[np.where((imagem_extraida == [0,0,0]).all(axis = 2))] = [0,255,0]

plt.figure(figsize=(20,10))
plt.imshow(imagem_extraida)
plt.title("Pirmide Extrada")
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
print(cv2.__version__)

%matplotlib inline

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
imagem = cv2.imread("imagens/nota-fiscal.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
height, width = imagem.shape[:2]

imagem_marcadores = imagem.copy()

pontos_marcadores = np.float32([[130,30],[400,30],[95,605],[440,605]])
pontos_destino = np.float32([[0,0],[600,0],[0,800],[600,800]])

color = (0, 0, 255)

cv2.circle(imagem_marcadores, (130,30), 5, color, 3)
cv2.circle(imagem_marcadores, (400,30), 5, color, 3)
cv2.circle(imagem_marcadores, (95,605), 5, color, 3)
cv2.circle(imagem_marcadores, (440,605), 5, color, 3)

plt.figure(figsize=(20,10))
plt.imshow(imagem_marcadores)
plt.title("Nota fiscal com Marcadores")
```

```python
matriz_perspectiva = cv2.getPerspectiveTransform(pontos_marcadores, pontos_destino)
imagem_transformada = cv2.warpPerspective(imagem,matriz_perspectiva,(600,800))

plt.figure(figsize=(10,10))
plt.imshow(imagem_transformada)
plt.title("Nota fiscal expandida")
```

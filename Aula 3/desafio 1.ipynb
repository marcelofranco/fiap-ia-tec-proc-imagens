```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
imagem_gray = cv2.imread("imagens/robocop.jpg", cv2.IMREAD_GRAYSCALE)
imagem_suavizada = cv2.GaussianBlur(imagem_gray,(3,3),0)
borda_canny = cv2.Canny(imagem_gray, 150, 150)
_, imagem_inverso = cv2.threshold(borda_canny, 30, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.figure(figsize=(15, 15))
plt.title("Borda Laplaciana")
plt.imshow(imagem_inverso, cmap="gray")
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread("imagens/captcha-2.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
```

```python
imagem_suavizada = cv2.blur(image, (5,5), 0)
plt.figure(figsize=(8, 10))
plt.title("Bordas")
plt.imshow(imagem_suavizada, cmap="gray")
```

```python
ret, imagem_limiarizada = cv2.threshold(imagem_suavizada,160, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8, 10))
plt.imshow(imagem_limiarizada, cmap="gray")
plt.title("Limiarização binária")
```

```python
kernel = np.ones((1,1), np.uint8)
image_dilated = cv2.erode(imagem_limiarizada, kernel, iterations=1)

plt.figure(figsize=(8, 10))
plt.title("Dilatação")
plt.imshow(image_dilated, cmap="gray")
```

```python
image_edged = cv2.Canny(image_dilated, 40, 190)

plt.figure(figsize=(20, 10))
plt.title("Dilatação")
plt.imshow(image_edged, cmap="gray")
```

```python
kernel_d = np.ones((1,1), np.uint8)
image_dilated = cv2.dilate(image_edged, kernel_d, iterations=1)

plt.figure(figsize=(8, 10))
plt.title("Bordas")
plt.imshow(image_dilated, cmap="gray")
```

```python
_, contornos_objetos, hierarchy_objetos = cv2.findContours(image_dilated, 
                                                           cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print("Contornos encontrados "+str(len(contornos_objetos)))

contours_filtered = []
for idx, contour in enumerate(contornos_objetos):
    area = cv2.contourArea(contour)
    if area > 30:
        contours_filtered.append(contour)
        (x, y, w, h) = cv2.boundingRect(contour)
print("Contornos encontrados "+str(len(contours_filtered)))

image_original = cv2.imread("imagens/captcha-2.jpg")

cv2.drawContours(image_original, contours_filtered, -1, (0,0,255), 1)

plt.figure(figsize=(20, 20))
plt.title("Contornos")
plt.imshow(image_original)
```
